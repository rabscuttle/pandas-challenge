```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data_df['Item Name'].unique()
unique_item_count = len(unique_items) #Count of unique items

avg_purchase_price = purchase_data_df["Price"].mean() #Average purchase price
avg_purchase_price = round(avg_purchase_price, 2)

num_purchases = len(purchase_data_df["Purchase ID"]) #Number of purchases

total_revenue = purchase_data_df["Price"].sum() #Total Revenue

#Create Summary Dataframe
purchasing_analysis_df = pd.DataFrame({
    "Number of Unique Items": [unique_item_count],
    "Average Price": ["$" + str(avg_purchase_price)],
    "Number of Purchases": [num_purchases],
    "Total Revenue": ["$" + str(total_revenue)]
})

purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop duplicates from main DF
unique_users_df = purchase_data_df.drop_duplicates(subset='SN', keep="last")

# Create gender count DF
count_genders_df = unique_users_df["Gender"].value_counts().rename_axis("Gender").to_frame("Total Count")

# Create gender pct DF
gender_pct_df = round(unique_users_df["Gender"].value_counts(normalize=True) * 100, 2)
gender_pct_df = gender_pct_df.rename_axis("Gender").to_frame("Percentage of Players")

# Merge DFs
player_gender_stats_df = pd.merge(count_genders_df, gender_pct_df, on = "Gender")
player_gender_stats_df.index.name = ""

player_gender_stats_df.style.format({"Percentage of Players": "{:,.2f}%"})
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins
bins = [0, 9, 14, 19 ,24, 29, 34, 39, 100]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Create/Calculate DFs
unique_users_df["Age Summary"]= pd.cut(unique_users_df["Age"], bins, labels=labels)
age_counts_df=unique_users_df["Age Summary"].value_counts().rename_axis('Age Summary').to_frame('Total Count')
total_players_by_age_df= age_counts_df["Total Count"].sum()
age_percent_df = unique_users_df["Age Summary"].value_counts(normalize=True) * 100
final_age_percent_df = age_percent_df.rename_axis("Age Summary").to_frame('Percentage of Players')
avg_by_age_df = age_counts_df['Total Count']/total_players_by_age_df

summary_avg_by_age_df = pd.DataFrame(avg_by_age_df)
renamed_age_percentage_df = summary_avg_by_age_df.rename(columns={'Total Count': 'Percentage of Players'})

# Merge purchase by age data
age_demo_df = pd.merge(age_counts_df, final_age_percent_df  , on = "Age Summary")
age_demo_df = age_demo_df.sort_values("Age Summary").style.format({"Percentage of Players": "{:,.2f}%"})
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Establish Bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#Create Age Summary
purchase_data_df["Age Summary"]= pd.cut(purchase_data_df["Age"], bins, labels=labels)

#Count Purchases
purchase_counts_df=purchase_data_df["Age Summary"].value_counts().rename_axis('Age Summary').to_frame('Purchase Count')

#Calculate Avg Purchase
average_by_age_df = purchase_data_df.groupby('Age Summary', as_index=False).Price.mean()
indexed_average_by_age_df = average_by_age_df.set_index('Age Summary')
indexed_average_by_age_df = indexed_average_by_age_df.rename(columns={"Price": "Average Purchase Price"})

#Calculate Total By Age
total_by_age_df = purchase_data_df.groupby('Age Summary', as_index=False).Price.sum()
indexed_total_by_age_df = total_by_age_df.set_index('Age Summary')
indexed_total_by_age_df = indexed_total_by_age_df.rename(columns={"Price": "Total Purchase Value"})

#Calculate Average Purchase per Person
average_per_person_by_age_df = indexed_total_by_age_df['Total Purchase Value']/age_counts_df['Total Count']
average_by_age_df = pd.DataFrame(average_per_person_by_age_df)
average_by_age_df.columns = ['Avg Total Purchase per Person']

#Merge DFs Into One
first_df = pd.merge(purchase_counts_df, indexed_average_by_age_df, on = "Age Summary")
second_df = pd.merge(first_df, indexed_total_by_age_df, on = "Age Summary")
final_merge_age_df = pd.merge(second_df, average_by_age_df, on = "Age Summary")
final_merge_age_df.sort_values("Age Summary").style.format({"Average Purchase Price": "${:,.2f}", "Total Purchase Value": "${:,.2f}", "Avg Total Purchase per Person": "${:,.2f}"})
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Purchase Count By Screen Name
purchase_count_by_sn = purchase_data_df["SN"].value_counts().rename_axis('Name').to_frame('Purchase Count')

#Total Purchase Value By Screen Name
total_groupedby_SN = purchase_data_df.groupby('SN', as_index=False).Price.sum()
total_purchase_value_sn = total_groupedby_SN.set_index('SN')
total_purchase_value_sn  = total_purchase_value_sn .rename(columns={"Price": "Total Purchase Value"})
total_purchase_value_sn.index.rename('Name',inplace=True)

# Calculate Average Purchase Price by Name
price_avg_sn = total_purchase_value_sn['Total Purchase Value']/purchase_count_by_sn['Purchase Count']
avg_per_person_sn = pd.DataFrame(price_avg_sn)
avg_per_person_sn.columns = ['Avg Purchase Price']
avg_per_person_sn.index.rename('Name', inplace=True)

#Merge
merge_SN_df = pd.merge(purchase_count_by_sn, avg_per_person_sn, on = "Name")
final_merge_df = pd.merge(merge_SN_df, total_purchase_value_sn, on = "Name")
final_merge_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get Item ID, Item Name, and Item Price
df = purchase_data_df[["Item ID", "Item Name", "Price"]]

total_purchase_value_df = df.groupby(["Item ID", "Item Name"], as_index=False).Price.sum()

item_purchase_count = purchase_data_df["Item ID"].value_counts().rename_axis('Item ID').to_frame('Purchase Count')

total_purchase_value_df = total_purchase_value_df.rename(columns={"Price": "Total Purchase Value"})
total_purchase_value_df = total_purchase_value_df.set_index('Item ID')


unique_items_df = purchase_data_df.drop_duplicates(subset='Item ID')
price_df = unique_items_df.set_index('Item ID')
price_df = price_df[['Price']]
price_df

merge1 = pd.merge(item_purchase_count, total_purchase_value_df, on = "Item ID")
merge2 = pd.merge(merge1, price_df, on = "Item ID")
merge2.sort_values("Purchase Count", ascending=False).style.format({"Price": "${:,.2f}", "Total Purchase Value": "${:,.2f}"})
```
